```python
test_data = pd.read_csv('/kaggle/input/crime-learn/crime_test.csv')
test_data
```

```python
test_data.shape
```

```python
train_data = pd.read_csv('/kaggle/input/crime-learn/crime_train.csv')
train_data
```

```python
train_data.shape
```

```python
test_data.corr()
```

```python
X_train =train_data.iloc[:,:-2].values
X_train
```

```python
y_train = train_data.iloc[:,-1].values
y_train
```

```python
X_test= test_data.iloc[:,:-1].values
X_test
```

```python
from sklearn.linear_model import LinearRegression  
reg = LinearRegression()
reg.fit(X_train, y_train)
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X_train, y_train)
```

```python
final = reg.predict(X_test)
final
```

```python
final_output = pd.DataFrame({'ID':test_data['ID'],'Prediction':np.array(final)})
```

```python
final_output.to_csv("crime_final_output.csv")
```
